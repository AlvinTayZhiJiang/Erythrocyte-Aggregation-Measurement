```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("lead 1 ecg.csv")
df = pd.DataFrame(df, columns = ["Time","Potential"])
display(df.head())

df.plot(
    x = "Time",
    y = "Potential",
    title = "ECG Waveform (Lead 1)",
    figsize = (12,8),
    color="red",
    grid = True
)

plt.xlabel('Time (Sec)')
plt.ylabel('Potential (mV)')
plt.show()
```
